```python
#while повторяет выполнение блока кода до тех пора, пока заданное условие остается истинным
x = 15
while x < 40:
     x += 2
     print(x)
    
```

```python
#Цикл for  итерирует по элементам последовательности (например, списку или строке), выполняя блок кода для каждого элемента
месяц = "Сентябрь"
for буква in месяц:
    print(буква)
```

```python
#if выполняет код, если условие истинное. else выполняет код, если предыдущее условие ложное
y= 95
if y > 90:
     print("y больше 90")
else:
    print ("y меньше 90")
```
