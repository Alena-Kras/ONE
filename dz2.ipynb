```python
books = {
    "Phyton Basics": 3,
    "Data Science": 2,
    "Machine Learning": 1
}
history = {}

while True:
      
        command = input("Введите команду (take, return, list, exit):")
        
        if command == "take":
            book = input("Введите название книги:")
            print(books[book], "запрошенной вами книги")
            if books[book] == 0:
                  print("Эти книги закончились")
                  continue
            books[book] -= 1
            print(f"Вы забрали книгу {book}, осталось {books[book]}")
            history.append({command: book})
        elif command == "return":
                book = input("Введите название книги:")
                print(books[book], "запрошенной вами книги")
                books[book] += 1
                print(f"Вы вернули книгу {book}, осталось {books[book]}")
                history.append({command: book})
        elif command == "list":
            for name, kolvo in books.items():
                  if kolvo != 0:
                    print(f'{name}: {kolvo}')
        

        elif command == "exit":
                print("Выход из программы")    
                break            
print(history)
```
